```python
X = parkinsons_data.drop(columns=['name','status'], axis=1)
Y = parkinsons_data['status']
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

```python
model = svm.SVC(kernel='linear')


# training the SVM model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)


print('Accuracy score of training data : ', training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)


print('Accuracy score of test data : ', test_data_accuracy)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd #Added this line to read in data file
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score


parkinsons_data = pd.read_csv('/content/parkinsons .csv')

X = parkinsons_data.drop(columns=['name', 'status'], axis=1)  # dropping 'name' and target
Y = parkinsons_data['status']

test_sizes = np.linspace(0.1, 0.5, 10)  # 10 points from 10% to 50%
accuracies = []

for test_size in test_sizes:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, stratify=Y, random_state=2)

    classifier = svm.SVC(kernel='linear')
    classifier.fit(X_train, Y_train)

    X_test_prediction = classifier.predict(X_test)
    acc = accuracy_score(Y_test, X_test_prediction)
    accuracies.append(acc)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(test_sizes, accuracies, marker='o', linestyle='-', color='royalblue')
plt.title('Accuracy vs Test Size (Parkinsons SVM Model)')
plt.xlabel('Test Size')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
